# Count all digits of a number

**Solution 1**

```python
def count_digits1(n):
  if n==0:
    return 1
  count=0
  while n!=0:
    n=n//10
    count+=1
  return count
```

```python
count_digits1(100)
```

**Solution 2**

```python
import math
```

```python
def count_digits(n):
  if n==0:
    return 1
  else:
    return math.floor(math.log10(n))+1
```

```python
count_digits(1000)
```

# Count Odd digits of a number

```python
def odd_counts(m):
  count=0

  while m!=0:
    last_digit = m%10
    if last_digit%2==1:
      count+=1
    m=m//10

  return count
```

```python
odd_counts(513)
```

# Reverse a number

```python
def reverse(n):
  num=0
  while n>0:
    last=n%10
    num=num*10+last
    n=n//10
  return num
```

```python
reverse(123456789)
```

# Palindrome number

```python
def palindrome(n):
  return n==reverse(n)
```

```python
palindrome(11011)
```

# Largest Digit

```python
def largest_digit(n):
  largest=0
  while n>0:
    digit=n%10
    if digit>largest:
      largest=digit
    n=n//10
  return largest
```

```python
largest_digit(11222000)
```

# Divisors of a Number

Solution 1: Brute Force

```python
def divisors(n):
  #created a list to store the answers
  divs=[]

  #in case of 1
  if n==1:
    divs.append(1)
    return divs

  #loop to find all the divisors
  for i in range(1, n+1):
    if n%i==0:
      divs.append(i)

  #return list
  return divs
```

```python
divisors(6)
```

Solution 2: optimal

```python
import math
```

```python
def divisors1(n):
  #created a list to store the answers
  divs=[]

  #in case of 1
  if n==1:
    divs.append(1)
    return divs

  #loop to find all the divisors
  for i in range(1, int(math.sqrt(n))+1):
    if n%i==0:
      divs.append(i)

      if n//i!=n%i and n//i!=i:
        divs.append(n//i)

  #return list
  return sorted(divs)
```

```python
divisors1(12)
```

```python
divisors1(9)
```

# Factorial

```python
def factorial(n):
  ans=1
  for i in range(1,n+1):
    ans=ans*i

  return ans
```

# Armstronge

```python
def armstronge(n):
  arm=0
  final=n
  count=int(math.log10(n))+1
  while n>0:
    last=n%10
    arm=arm+(last**count)
    n=n//10

  return final==arm
```

```python
armstronge(153)
```

# Perfect Number

**Solution 1: naive**

```python
def perfect_number(n):
  sum=0
  for i in range(1,n-1):
    if n%i==0:
      sum+=i

  return n==sum
```

```python
perfect_number(6)
```

**Solution 2: Best**

```python
def perfect(n):
  #variable to calculate sum
  sum=0

  #Loop till int(square root of n), in case of 6 it will run till 1 to 2
  for i in range(1,int(math.sqrt(n))+1):

    #Condition of divisor
    if n%i==0:
      #add sum
      sum=sum+i

      #Other divisor of the number i
      if n//i !=i and n//i!=n:
        #add sum
        sum=sum+(n//i)

  #retrun bolean value of comparison between sum and n
  return sum==n
```

```python
perfect(8128)
```

# Prime Number

Solution 1: Brute Force

```python
def prime_number(n):
  #checking for 1
  if n==1:
    return False

  #loop for other numbers
  for i in range(2,n-1):
    if n%i==0:
      return False

  return True
```

```python
prime_number(97)
```

# Prime Numbers till N

```python
def count_prime(n):
  #variable for count
  count=0

  #using our previously created prime function
  for i in range(2,n+1):
    if primenumber(i)==True:
      count+=1

  return count
```

# GCD of two numbers

Solution: Brute Force

```python
def GCD(n1,n2):
  l1=divisors1(n1)
  l2=divisors1(n2)
  div=0

  for i in range(len(l1)-1,-1,-1):
    if l1[i] in l2:
      if l1[i]>div:
        div = l1[i]
    else:
      i=i-1

  return div
```

```python
GCD(6,12)
```

Solution: Optimal

```python
def GCD_1(n1,n2):
  #looping
  while n1!=0 and n2!=0:
    if n1>n2:
      n1=n1%n2
    else:
      n2=n2%n1

  return max(n1,n2)
```

```python
GCD_1(6,12)
```

# LCM of two numbers

Solution: Brute Force

```python
def LCM_1(n1,n2):
  #creating variable
  max1=max(n1,n2)

  #if n1 and n2 are same
  if n1==n2:
    return n1

  #looping to find the largest multiplication possible which is divisable by both
  for i in range(1,max1):
    lcm=max1*i
    if lcm%n1==0 and lcm%n2==0:
      return lcm
```

```python
LCM_1(14,15)
```

```python
def LCM(n1, n2):
  # Variable to store lcm
  lcm = 0

  # Variable to store max of n1 & n2
  n = max(n1, n2)
  i = 1

  while True:
    # Variable to store multiple
    mul = n * i

    # Checking if multiple is common
    # common for both n1 and n2
    if mul % n1 == 0 and mul % n2 == 0:
        lcm = mul
        break
    i += 1

  # Return the stored LCM
  return lcm
```

```python
LCM(14,15)
```

```python
def LCM_2(n1,n2):
  return n1*n2//GCD_1(n1,n2)
```
